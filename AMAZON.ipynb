```python
from pandas_datareader import data
import pandas as pd
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

```python
# 取得する日の範囲を指定する
start = datetime.datetime(2020, 1, 1)
end = datetime.datetime(2020, 12, 31)

f = data.DataReader('AMZN', 'yahoo', start, end)

plt.title('AMAZON from 2020.1.1 to 2020.12.31')

# fill_between でその日の最高値と最低値をプロットする
plt.fill_between(f.index, f['Low'], f['High'], color="b", alpha=0.5)

# 始値をプロットする。
# 自動的に Index が Date になっているので、横軸が時間になる。
plt.figure(figsize=(30,20))

f['Open'].plot()
plt.show()
```
